```python
%matplotlib inline
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os


# Load CSV
file_to_read_city = os.path.join("resources", "city_data.csv")
city_data_df = pd.read_csv(file_to_read_city)
city_data_df.head(10)
```

```python
#Load CSV
file_to_read_ride = os.path.join("resources", "ride_data.csv")
ride_data_df = pd.read_csv(file_to_read_ride)
ride_data_df.head(10)
```

```python
# Get the columns and the rows that are not null.
ride_data_df.count()
```

```python
# Get the columns and the rows that are not null.
ride_data_df.isnull().sum()
```

```python
# Get the data types of each column.
ride_data_df.dtypes
```

```python
# Get the unique values of the type of city.
city_data_df["type"].unique()
```

```python
# Get the number of data points from the Urban cities.
sum(city_data_df["type"]=="Urban")
```

```python
# Get the number of data points from the Suburban cities.
sum(city_data_df["type"]=="Suburban")
```

```python
# Get the number of data points from the Rural cities.
sum(city_data_df["type"]=="Rural")
```

```python
# Merge Data Frames
pyber_data_df = pd.merge(ride_data_df, city_data_df, on=["city", "city"], how='left')
pyber_data_df.head()
```

```python
# Create the Urban city DataFrame.
urban_cities_df = pyber_data_df[pyber_data_df["type"] == "Urban"]
urban_cities_df.head()
```

```python
# Create the Suburban and Rural city DataFrames.
suburban_cities_df = pyber_data_df[pyber_data_df["type"] == "Suburban"]
rural_cities_df = pyber_data_df[pyber_data_df["type"] == "Rural"]
```

```python
rural_cities_df.head()
```

```python
urban_ride_count = urban_cities_df.groupby(["city"]).count()["ride_id"]
urban_ride_count.head()
```

```python
# Create the suburban and rural ride count.
suburban_ride_count = suburban_cities_df.groupby(["city"]).count()["ride_id"]

rural_ride_count = rural_cities_df.groupby(["city"]).count()["ride_id"]
```

```python
# average fare for each city
urban_avg_fare = urban_cities_df.groupby(["city"]).mean()["fare"]
suburban_avg_fare = suburban_cities_df.groupby(["city"]).mean()["fare"]
rural_avg_fare = rural_cities_df.groupby(["city"]).mean()["fare"]

urban_avg_fare.head()
```

```python
# Get the average number of drivers for each city.
urban_driver_count = urban_cities_df.groupby(["city"]).mean()["driver_count"]
suburban_driver_count = suburban_cities_df.groupby(["city"]).mean()["driver_count"]
rural_driver_count = rural_cities_df.groupby(["city"]).mean()["driver_count"]
```

```python
# Build the scatter plots for urban cities.
plt.scatter(urban_ride_count, 
            urban_avg_fare, 
            label="Urban",
           s=10*urban_driver_count,
           c="coral",
           edgecolor="black", linewidths=1,
           alpha=0.8)
plt.title("Pyber Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)
plt.legend()
```

```python
# Build the scatter plots for suburban cities.
plt.scatter(suburban_ride_count, 
            suburban_avg_fare, 
            label="Suburban",
           s=10*suburban_driver_count,
           c="skyblue",
           edgecolor="black", linewidths=1,
           alpha=0.8)
plt.title("Pyber Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)
plt.legend()
```

```python
# Build the scatter plots for rural cities.
plt.scatter(rural_ride_count, 
            rural_avg_fare, 
            label="rural",
           s=10*rural_driver_count,
           c="gold",
           edgecolor="black", linewidths=1,
           alpha=0.8)
plt.title("Pyber Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)
plt.legend()
```

```python
# Combine all scatter plots
plt.subplots(figsize=(10,6))
plt.scatter(urban_ride_count, 
            urban_avg_fare, 
            label="Urban",
           s=10*urban_driver_count,
           c="coral",
           edgecolor="black", linewidths=1,
           alpha=0.8)

plt.scatter(suburban_ride_count, 
            suburban_avg_fare, 
            label="Suburban",
           s=10*suburban_driver_count,
           c="skyblue",
           edgecolor="black", linewidths=1,
           alpha=0.8)

plt.scatter(rural_ride_count, 
            rural_avg_fare, 
            label="rural",
           s=10*rural_driver_count,
           c="gold",
           edgecolor="black", linewidths=1,
           alpha=0.8)

# Incorporate the other graph properties
plt.title("PyBer Ride-Sharing Data (2019)", fontsize=20)
plt.ylabel("Average Fare ($)", fontsize=12)
plt.xlabel("Total Number of Rides (Per City)", fontsize=12)
plt.grid(True)

# Add the legend.
plt.legend()
# Show the plot
plt.show()
```
